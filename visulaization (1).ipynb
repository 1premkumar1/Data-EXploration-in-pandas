### Distplot is created for the variable "sales" where the data has its peak value from 10000 to 20000, the value decrease on the range of 20000 to 40000.

```python
plt.scatter(softdrinks['sales'],softdrinks['quantity'])
plt.xticks(color='red',fontweight='bold',fontsize='12',rotation='90')
plt.show()
```

### In this scatterplot, variables ''sales'' and ''quantity'', there is outlier lies in 70000 in yaxis, the maximum value of data in the dataset is occured between the values of (5000 to 45000) in x-axis.

```python
x=list(softdrinks.iloc[:3])
y=list(softdrinks.iloc[:4])
plt.bar(x,y,color='r')
plt.title("sales vs quantity")
plt.xlabel('sales')
plt.ylabel('quantity')
plt.xticks(color='r',fontweight='bold',fontsize='12',rotation='85',horizontalalignment='center')
```

### We have created a subplots for two varivables "sales,quantity"

```python
sns.jointplot('sales','quantity',softdrinks)
plt.show()
```

### In this joinplot we have chosen a two variables "sales" and "quantity" 

```python
softdrinks=softdrinks[(softdrinks.sales<15000)&(softdrinks.quantity<30000)]
sns.jointplot('sales','quantity',softdrinks)
plt.show()
```

### In this joinplot we have chosen a certain value and condition for the two variables "sales"<15000 and "quantity"<30000.

```python
softdrinks=softdrinks[(softdrinks.sales<15000)&(softdrinks.quantity<30000)]
sns.jointplot('sales','quantity',softdrinks)
sns.jointplot(x='sales',y='quantity',data=softdrinks,kind='hex',color='r')
plt.show()
```
